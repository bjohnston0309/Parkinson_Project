```python
import csv
import PyQt5
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import pylatex as tex

#Importing data from csv file and setting variables
data=np.genfromtxt("../Example Data/synthetic_eyemovement_001_1kHz.csv", delimiter=',')
data1=data.transpose()
t=(data1[0])
x=(data1[1])
y=(data1[2])

# Plotting t-x and t-y data for synthetic data set
figure, axis = plt.subplots(2)
figure.tight_layout(pad=2.8)
axis[0].plot(t,x)
axis[0].set_title('$x$-$t$ data')
axis[0].set_xlabel('$t$ $(s)$')
axis[0].set_ylabel('$x$ $(arcmin)$')
axis[1].plot(t,y)
axis[1].set_title('$y$-$t$ data')
axis[1].set_xlabel('$t$ $(s)$')
axis[1].set_ylabel('$y$ $(arcmin)$')
```

```python
#Plotting x-y data 
plt.plot(x,y)
plt.title('$y$-$x$ data')
plt.xlabel('$x$ $(arcmin)$')
plt.ylabel('$y$ $(arcmin)$')
plt.show()
```
