```python
import numpy as np
import pandas as pd
import os
import math
import matplotlib.pyplot as plt

t = np.linspace(0,20,20)

x0_1 = 0.31
x0_2 = 0.3105
y0_1 = 0.01
a = 1.4
b = 0.3
x1 = []
x2 = []

for i in range(len(t)):

    x1.append(1 - (a*x0_1)**2 + y0_1)
    y1 = b*y0_1

    x0_1 = x1[i]
    y0_1 = y1

for i in range(len(t)):

    x2.append(1 - (a*x0_2)**2 - y0_1)
    y1 = b*y0_1

    x0_2 = x2[i]
    y0_1 = y1




plt.plot(t,x1, color='tab:blue')
plt.plot(t,x2,'--', color='tab:blue')
plt.xlabel('t', fontsize=16)
plt.ylabel('x', fontsize=16)
plt.legend(['x\u2080 = 0.3100','x\u2080 = 0.3105'])
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)





```
